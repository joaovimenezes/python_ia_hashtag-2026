### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

Arquivos da aula: https://drive.google.com/drive/folders/1FbDqVq4XLvU85VBlVIMJ73p9oOu6u2-J?usp=drive_link

```python
import pandas as pd

df = pd.read_csv('clientes.csv')

display(df)

display(df.info())
```

```python
from sklearn.preprocessing import LabelEncoder as le

#------ Codificando os textos para números (integração com a IA) ----------
codificador_profissao = le() 
df['profissao'] = codificador_profissao.fit_transform(df['profissao'])

codificador_mix = le()
df['mix_credito'] = codificador_mix.fit_transform(df['mix_credito'])

codificador_comportamento = le()
df['comportamento_pagamento'] = codificador_comportamento.fit_transform(df['comportamento_pagamento'])

display(df)

```

```python
#-------- Separando o Y e o X para o treinamento do ML ----------------
y = df['score_credito']

x = df.drop(columns={'score_credito', 'id_cliente'})

from sklearn.model_selection import train_test_split as tts

#-------- Dividindo automaticamente o Y e X para teste ----------------
x_train, x_test, y_train, y_test = tts(x,y, test_size=0.3)
```

```python
#--------- Importando dois modelos de ml: KNeighbors e RandomForest ---------------
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn.ensemble import RandomForestClassifier as rf

modelo_knn = knn()
modelo_rf = rf()

#--------- Treinando os dois modelos com o Y e X de treino -------------------------
modelo_knn.fit(x_train,y_train)
modelo_rf.fit(x_train,y_train)
```

```python
#--------- Fazendo o teste da previsão dos modelos com o X teste ------------------
previsao_knn = modelo_knn.predict(x_test)
previsao_rf = modelo_rf.predict(x_test)

from sklearn.metrics import accuracy_score as asc

#---------- Verificando a precisão dos modelos de acordo com o gabarito -----------
print(asc(y_test, previsao_knn))
print(asc(y_test, previsao_rf))
```

```python
novos_clientes = pd.read_csv('novos_clientes.csv')

display(novos_clientes)

#---------------- Utilizando os codificadores na tabela de novos clientes --------------
novos_clientes['profissao'] = codificador_profissao.transform(novos_clientes['profissao'])
novos_clientes['mix_credito'] = codificador_mix.transform(novos_clientes['mix_credito'])
novos_clientes['comportamento_pagamento'] = codificador_comportamento.transform(novos_clientes['comportamento_pagamento'])

#--------------- Fazendo a previsão do Score de Crédito ---------------------------------
previsao = modelo_rf.predict(novos_clientes)

display(previsao)

```
